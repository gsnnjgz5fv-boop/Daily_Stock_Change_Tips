```python
import akshare as ak
import schedule
import time
from datetime import datetime

def send_stock_data():
    stock_data = ak.stock_zh_a_spot()
    target_codes = ['600519', '000858']
    result = stock_data[stock_data['代码'].isin(target_codes)][['代码','名称','最新价','涨跌幅']]
    
    # 推送逻辑（示例：保存到文件/发送邮件）
    with open("result.md", "w") as f:
        f.write(result.to_markdown(index=False))
    print(f"{datetime.now()} 数据已更新")

# 设置每天14:30执行
schedule.every().day.at("14:30").do(send_stock_data)

while True:
    schedule.run_pending()
    time.sleep(60)
```
